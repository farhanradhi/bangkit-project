```python
import csv
import textwrap

# Fungsi untuk merapikan tulisan menjadi paragraf
def format_output_as_paragraph(content, line_width=80):
    return "\n".join(textwrap.wrap(content, width=line_width))

# Fungsi untuk membaca data dari file CSV
def read_csv_data(filename):
    data = {
        "topic": [],
        "questions": [],
        "answers": []
    }
    with open(filename, mode='r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        current_category = None
        questions = []
        answers = []

        for row in reader:
            if row["Kategori"] != current_category:
                # Simpan kategori sebelumnya
                if current_category is not None:
                    data["questions"].append(questions)
                    data["answers"].append(answers)

                # Reset untuk kategori baru
                current_category = row["Kategori"]
                data["topic"].append([current_category])
                questions = []
                answers = []

            # Tambah pertanyaan dan jawaban
            questions.append(row["Pertanyaan"])
            answers.append(row["Jawaban"])

        # Tambahkan kategori terakhir
        if current_category is not None:
            data["questions"].append(questions)
            data["answers"].append(answers)

    return data

# Fungsi untuk menampilkan daftar topik
def display_topics(data):
    print("\nDaftar Topik:")
    for index, topic in enumerate(data["topic"]):
        print(f"{index + 1}. {topic[0]}")

# Fungsi untuk menampilkan deskripsi topik
def get_content(index, data):
    return format_output_as_paragraph(f"Topik: {data['topic'][index][0]}")

# Fungsi untuk menampilkan daftar pertanyaan
def display_questions(index, data):
    print("\nDaftar Pertanyaan:")
    for qid, question in enumerate(data["questions"][index]):
        print(f"{qid + 1}. {question}")

# Fungsi untuk mendapatkan jawaban dari pertanyaan yang dipilih
def get_answer(question, index, data, previous_answers):
    # Cek pertanyaan ada kata kunci pengertian atau definisi
    if "pengertian" in question.lower() or "definisi" in question.lower():
        return get_content(index, data)

    # Cari jawaban berdasarkan input pertanyaan
    for qid, predefined_question in enumerate(data["questions"][index]):
        if question.lower() in predefined_question.lower():
            answer = data["answers"][index][qid]
            answer_formatted = format_output_as_paragraph(answer)
            if answer_formatted not in previous_answers:
                previous_answers.add(answer_formatted)
                return answer_formatted

    # Jika tidak ditemukan jawaban
    return "Maaf, saya tidak memiliki informasi tentang itu."

# Fungsi utama
def main():
    # File CSV di Google Drive
    filename = '/content/drive/Shareddrives/Dataset/datakedua/toptwenty.csv'  # Sesuaikan dengan lokasi file CSV Anda

    # Membaca data dari file CSV
    data = read_csv_data(filename)

    while True:
        print("\n" + "=" * 50)
        display_topics(data)
        choice = input("\nPilih topik (masukkan nomor atau 'q' untuk keluar): ")

        if choice.lower() == 'q':
            print("\nTerima kasih! Sampai jumpa!")
            break

        if choice.isdigit() and 1 <= int(choice) <= len(data["topic"]):
            index = int(choice) - 1
            print("\n" + "=" * 50)
            print("\nDeskripsi Topik:")
            print(get_content(index, data))
            display_questions(index, data)

            # Inisialisasi set untuk menyimpan jawaban yang telah diberikan
            previous_answers = set()

            while True:
                question_choice = input("\nPilih pertanyaan (masukkan nomor), tulis pertanyaan Anda sendiri, atau ketik 'b' untuk kembali: ")

                if question_choice.lower() == 'b':
                    break

                if question_choice.isdigit() and 1 <= int(question_choice) <= len(data["questions"][index]):
                    q_index = int(question_choice) - 1
                    answer = data["answers"][index][q_index]
                    print("\n" + "=" * 50)
                    print(f"\nJawaban:\n{format_output_as_paragraph(answer)}\n")
                else:
                    # Pass previous_answers ke get_answer
                    answer = get_answer(question_choice, index, data, previous_answers)
                    print("\n" + "=" * 50)
                    print(f"\nJawaban:\n{answer}\n")
        else:
            print("Pilihan tidak valid. Silakan coba lagi.")

if __name__ == "__main__":
    main()
```
